```python
import pandas as pd
import matplotlib.pyplot as plt
#data importing
data=pd.read_csv("data.csv")
```

```python
plt.figure(figsize=(12,8))
plt.title("Momentum")
plt.plot(data["Date"], data["mom_ret_SPY"], label="SPY")
plt.plot(data["Date"], data["mom_ret_BND"], label="BND")
plt.plot(data["Date"], data["mom_ret_EURUSD=X"], label="EUROUSD=X")
plt.plot(data["Date"], data["mom_ret_GLD"], label="GLD")
plt.plot(data["Date"], data["mom_ret_USO"], label="USO")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Volatility")
plt.plot(data["Date"], data["vol_ret_SPY"], label="SPY")
plt.plot(data["Date"], data["vol_ret_BND"], label="BND")
plt.plot(data["Date"], data["vol_ret_EURUSD=X"], label="EUROUSD=X")
plt.plot(data["Date"], data["vol_ret_GLD"], label="GLD")
plt.plot(data["Date"], data["vol_ret_USO"], label="USO")
plt.legend()
plt.show()
```

```python
returns=data.loc[:,"vol_ret_BND": "vol_ret_USO"]
corr_matrix=returns.corr()
import seaborn as sns
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
returns=data.loc[1:500,"ret_BND": "vol_ret_USO"]

from sklearn.decomposition import PCA
transformer=PCA(n_components=0.95) #PCA transformation
components=transformer.fit_transform(returns) #component matrix obtained from PCA transformation

print(transformer.explained_variance_ratio_)
print(transformer.components_) #linear weight assigned to each input variable to create the single component
print(components.shape)  
```
